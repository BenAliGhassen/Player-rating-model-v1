```python
df = df[df['minute'] < 45]


player_pos = df[['player_name','position_name','team_name']].drop_duplicates()

player_pos
```

```python
player_stats = {}
for player_name in df['player_name'].unique():
    
    player = df[df['player_name'] == player_name]
    
    name = player['player_name']
    goals = len(player[player['outcome_name'] == 'Goal'])
    pass_count = len(player[player['outcome_name'].isnull()])
    assists = len(player[player['pass_goal_assist'] == True])
    minutePlayed = player.minute.max()
    fouls = player.foul_committed_offensive.count()
    saves = player.shot_deflected.count()
    

    player_stats[player_name] = {
        'Goals': (goals / minutePlayed) * 90,
        'Assists': (assists / minutePlayed) * 90,
        'Pass_count' : pass_count,
        'Minutes played' : minutePlayed,
        'Fouls' : fouls,
        'Saves' : saves,
        

    }


print(player_stats)
```

```python
Player_Rate = {}
for i in player_stats : 
    Player_Rate[i] = {
        'Rate' : 6 + 1.1 * player_stats[i]['Goals'] + .5 * player_stats[i]['Assists'] - .3 * player_stats[i]['Fouls'] + .008 * player_stats[i]['Pass_count'] + .05 * player_stats[i]['Saves'],
        
    }



Player_Rate
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

# Example dataset (replace with your actual dataset)
data = pd.DataFrame({
    'player_name': [
        'Marcos Javier Acuña', 'Nicolás Hernán Otamendi', 'Enzo Fernandez', 
        'Lionel Andrés Messi Cuccittini', 'Alejandro Darío Gómez', 
        'Cristian Gabriel Romero', 'Nahuel Molina Lucero', 'Rodrigo Javier De Paul', 
        'Julián Álvarez', 'Damián Emiliano Martínez', 'Alexis Mac Allister'
    ],
    'position_name': [
        'Left Back', 'Left Center Back', 'Center Defensive Midfield', 
        'Center Forward', 'Left Wing', 'Right Center Back', 'Right Back', 
        'Right Center Midfield', 'Right Wing', 'Goalkeeper', 'Left Center Midfield'
    ],
    'Rate': [7.0, 8.0, 7.5, 9.0, 8.0, 7.5, 7.2, 7.8, 8.2, 7.9, 7.4]
})

# Define x, y coordinates for positions
position_coordinates = {
    'Goalkeeper': (5, 40),
    'Left Back': (20, 10),
    'Left Center Back': (20, 30),
    'Right Center Back': (20, 50),
    'Right Back': (20, 70),
    'Center Defensive Midfield': (45, 40),
    'Left Center Midfield': (60, 30),
    'Right Center Midfield': (60, 50),
    'Left Wing': (80, 20),
    'Right Wing': (80, 60),
    'Center Forward': (95, 40)
}

# Map positions to coordinates
data['x'] = data['position_name'].map(lambda pos: position_coordinates[pos][0])
data['y'] = data['position_name'].map(lambda pos: position_coordinates[pos][1])

# Create subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

# Plot Messi image on the first subplot
img = imread('./messi.png') 
axs[0].imshow(img)
axs[0].axis('off')  
axs[0].set_title("MVP")

# Create the pitch on the second subplot
pitch = Pitch(pitch_type='statsbomb', line_color='white', pitch_color='grass')
pitch.draw(ax=axs[1])  # Draw the pitch on the second subplot

# Plot players on the pitch
for _, row in data.iterrows():
    # Add player name
    axs[1].text(row['x'], row['y'], row['player_name'], ha='center', va='center', fontsize=10, color='white',
                bbox=dict(facecolor='blue', edgecolor='black', boxstyle='round,pad=0.3'))
    # Add player rating next to the name (adjusted for better visibility)
    axs[1].text(row['x'], row['y'] - 5, f"{row['Rate']:.1f}", ha='center', va='center', fontsize=15, color='red')

# Set title for the second subplot
axs[1].set_title("Argentina Players Ratings")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
